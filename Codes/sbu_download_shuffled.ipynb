```python
# Paths to your URL file and the folder to save images
url_file = 'SBU_captioned_photo_dataset_urls.txt'
download_folder = 'sbu_downloaded'
```

```python
# Create the download folder if it doesn't exist
os.makedirs(download_folder, exist_ok=True)
```

```python
# Function to download an image and save it with the original order ID
def download_image(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(os.path.join(download_folder, filename), 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {filename}")
        return True
    else:
        print(f"Failed to download: {filename}")
        return False
```

```python
# Read the list of URLs into a pandas DataFrame
df = pd.read_csv(url_file, names=['URL'])

# Add a column for the original index
df['OriginalIndex'] = range(len(df))
```

```python
# df['URL'][444936]
```

```python
# Shuffle the DataFrame to randomize the order
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python

# Download 8000 valid images with their original order ID as filenames
valid_count = 0
for _, row in df.iterrows():
    if valid_count >= 8000:
        break
    url = row['URL']
    original_index = row['OriginalIndex']
    filename = f"{original_index}.jpg"
    if download_image(url, filename):
        valid_count += 1

print(f"Downloaded {valid_count} valid images to {download_folder}")
```
